## Setup and Loading Data

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
train = pd.read_csv("Train.csv", index_col="Applicant_ID")

train.head(2)
```

```python
test = pd.read_csv("Test.csv", index_col="Applicant_ID")

test.head(2)
```

### Checking & Encoding Categorical Columns

```python
dict_ = {'charge': 1,
         'lending': 0}

train.form_field47 = train.form_field47.map(dict_)
test.form_field47 = test.form_field47.map(dict_)

train.default_status = train.default_status.map({'no': 0, 'yes': 1})
```

```python
train.head(2)
```

## Feature Engineering

```python
train[['form_field32', 'form_field33']]
```

### Separating the Target from the predictors

```python
y = train.pop("default_status")

train.head(2)
```

### Handling Missing Values

```python
from sklearn.impute import SimpleImputer

zero_imputer = SimpleImputer(strategy='constant', add_indicator=True)
constant_imputer = SimpleImputer(strategy='constant', fill_value=-999)
```

```python
imputed_train = zero_imputer.fit_transform(train)
imputed_test = zero_imputer.transform(test)
```

## Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_mm = MinMaxScaler()

scaled_train = pd.DataFrame(scaler_mm.fit_transform(imputed_train))
scaled_test = pd.DataFrame(scaler_mm.transform(imputed_test))
```

## Modelling

### Local Validation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
params = {
    'n_estimators': 5000,
    'learning_rate': 0.005,
    'max_depth': 8,
    'objective': 'CrossEntropy',
    'eval_metric':'AUC',
    'random_seed': 3500,
    'early_stopping_rounds': 200,
    'use_best_model': True,
    'thread_count': -1,
}
```

```python
model1 = CatBoostClassifier(**params)
model2 = CatBoostClassifier(verbose=False)
model3 = CatBoostClassifier(verbose=False)
```

```python
kf = StratifiedKFold(n_splits=25, shuffle=True, random_state=99)
```

```python
# local_validation = []
test_predictions = []

for i, (train_index, val_index) in enumerate(kf.split(scaled_train, y)):
    print("*" * 75 + "\nFold" + str(i))
    cv_train, cv_val = scaled_train.iloc[train_index], scaled_train.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]
    
    model1.fit(cv_train, y_train, eval_set=[(cv_val, y_val)], verbose=100)
    y_pred = model1.predict_proba(cv_val)[:, 1]
    local_validation.append(model1.get_best_score()['validation']['AUC'])
    
    test_pred = model1.predict_proba(scaled_test)[:, 1]
    test_predictions.append(test_pred)   
```

```python
print(local_validation)
np.mean(local_validation)
```

```python
test_pred = np.mean(test_predictions, axis=0)
```

### Final Steps

```python
submission = pd.read_csv("SampleSubmission.csv")
```

```python
submission.default_status = test_pred
submission.to_csv("sub_stratified", index=False)
```
